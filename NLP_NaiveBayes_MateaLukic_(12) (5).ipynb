```python

import requests
from bs4 import BeautifulSoup
main_urls=["https://sportske.jutarnji.hr/sn/kosarka/",'https://sportske.jutarnji.hr/sn/nogomet/sp2022', "https://sportske.jutarnji.hr/sn/sport-mix/rukomet", "https://sportske.jutarnji.hr/sn/sport-mix/vaterpolo"]

urls = []
for url in main_urls:
  reqs = requests.get(url)
  soup = BeautifulSoup(reqs.text, 'html.parser')
  
  
  for link in soup.find_all('a'):
      urls.append(link.get('href'))
      print(link.get('href'))
```

```python
links=[]
cnt=0
for url in urls:
    cnt+=1
    if(cnt%2==0):
      if(str(url).startswith("/sn/nogomet/sp2022/")  or str(url).startswith("/sn/kosarka/") or str(url).startswith("/sn/sport-mix/rukomet/") or str(url).startswith("/sn/sport-mix/vaterpolo/")):
        url1="https://sportske.jutarnji.hr" + url 
        print(url1)
        links.append(url1)
```

```python
contents=[]
import requests
from bs4 import BeautifulSoup
# Make a request
for link in links:
    page = requests.get(str(link))
    soup = BeautifulSoup(page.content, 'html.parser')
    content= []
    for el in soup.select('p'):
       #print(el.text)
       content.append(el.text)
    contents.append(content[6:])
    


```

```python
for content in contents:
  print(content)
```

```python
contents2=[]
for content in contents:
  res = " ".join([str(item) for item in content])
  contents2.append(res)
```

# Sakupljanje korpusa za 3. klasu

```python
import requests
from bs4 import BeautifulSoup
main_urls=["https://www.bug.hr/znanost", "https://www.bug.hr/biznis","https://www.bug.hr/appdana","https://www.bug.hr/recenzije"]

urls = []
for url in main_urls:
  reqs = requests.get(url)
  soup = BeautifulSoup(reqs.text, 'html.parser')
  
  
  for link in soup.find_all('a'):
        if(type(link.get('href')) is str):
          new= link.get('href')
          urls.append(new)
          print(new)
```

```python
links=[]
cnt=0
for url in urls:
    cnt+=1
    if(cnt%3==0 and len(url)>30):
      if(str(url).startswith("/znanost") or str(url).startswith("/biznis") or str(url).startswith("/appdana")):
          new = "https://www.bug.hr" + url
          print(new)
          links.append(new)
```

```python
contents=[]
# Make a request
for link in links:
    page = requests.get(str(link))
    soup = BeautifulSoup(page.content, 'html.parser')
    content= []
    for el in soup.select('p'):
       #print(el.text)
       content.append(el.text)
    contents.append(content[1:6])

for content in contents:
   print(content)
```

```python
contents2=[]
for content in contents:
  res = " ".join([str(item) for item in content])
  contents2.append(res)

print(len(contents2))

for con in contents2:
  print(con)
```

```python
klasa3=""
for con in contents2:
  klasa3+=con + " "

  # prebroji rijeci (pojavnice)
words = klasa3.split()

print('Number of words in text file :', len(words))
```

```python
#open text file
text_file = open("znanost.txt", "w")
 
#write string to file
text_file.write(klasa2)
 
#close file
text_file.close()
```
